# L0 approximation with different values of sigma
For large sigma values, the approximation tends to L2, and as sigma shrinks it tends to L0.

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.cm as cm

delta = 0.025
x = np.arange(-1.0, 1.0, delta)
y = np.arange(-1.0, 1.0, delta)
X, Y = np.meshgrid(x, y)

plt.figure(figsize=(10,5))
for i, sigma in enumerate([0.001, 0.01, 0.1, 1, 10, 100]):
    Z = X**2 / (X**2+sigma) +Y**2 / (Y**2 + sigma)
    plt.subplot(2,4,i+1)
    plt.contour(X, Y, Z)
    plt.colorbar()
    plt.axis('equal')
    plt.title("sigma: " + str(sigma))
```

One-dimensional dependency on sigma

```python
sigma = np.arange(0.001, 20, 0.01)
x = 1
l0 = x**2 / (x**2 + sigma)

plt.plot(sigma, l0)
```

# Difference with tanh

```python
x = np.arange(-1, 1, 0.01)
sigma=0.1
sigma2=0.15
l0_root = x / np.sqrt(x**2 + sigma**2)

plt.plot(x, l0_root, 'r--')
plt.plot(x, l0_root**2, 'r')
plt.plot(x, np.tanh(x/sigma2), 'g')
plt.plot(x, np.tanh(x/sigma2)**2, 'g--')
```
